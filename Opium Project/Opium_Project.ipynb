```python
from google.colab import files
import pandas as pd

uploaded = files.upload()
file_name = next(iter(uploaded))

df = pd.read_excel(file_name, sheet_name=None)
```

```python
df_main_info = pd.read_excel(file_name, sheet_name='Main Info')
```

```python
df_alcohol_history = pd.read_excel(file_name, sheet_name='Alcoho Used')
df_opium_history = pd.read_excel(file_name, sheet_name='Opiume History')
df_Animal_Contact = pd.read_excel(file_name, sheet_name='Animal Contact')
```

```python
df_drug_history = pd.read_excel(file_name, sheet_name='DrugUsed')
```

```python
df_cancer_history = pd.read_excel(file_name, sheet_name='Family Cancer')
df_heating_systems = pd.read_excel(file_name, sheet_name='Heating Systmes')
```

```python
# Merge the DataFrames based on the patient ID (PID)
merged_df = df_main_info.merge(df_alcohol_history, on='PID', how='left')
merged_df = merged_df.merge(df_opium_history, on='PID', how='left')
merged_df = merged_df.merge(df_Animal_Contact, on='PID', how='left')
merged_df = merged_df.merge(df_drug_history, on='PID', how='left')
merged_df = merged_df.merge(df_cancer_history, on='PID', how='left')
merged_df = merged_df.merge(df_heating_systems, on='PID', how='left')
```

```python
merged_df.to_csv('Merged_dataset.csv', index=False)
```

```python
# Handling missing values
merged_df.fillna(method='ffill', inplace=True)  # Forward fill missing values
merged_df.dropna(inplace=True)  # Drop rows with any remaining missing values

# Remove duplicates based on the patient ID
merged_df.drop_duplicates(subset='PID', keep='first', inplace=True)
```
